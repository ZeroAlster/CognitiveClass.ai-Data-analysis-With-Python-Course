```python
'''
we want to detect missing values in every parts of our dataset
'''

missing_data=df.isnull()
missing_data.head(30)

'''
places where are true mean that the value is missed
'''

```

```python
'''
Count missing values in each column  using for loop
'''

for col in missing_data.columns.values.tolist():
  print(col)
  print(missing_data[col].value_counts())
  print("*"*20)
```

```python
'''
now we want to replace or remove missing data
'''


'''
calculate the avg of that column that has missing values
'''
avg1=df["normalized-losses"].astype("float").mean(axis=0)
avg1
```

```python
'''
Now we want to replace the missing vals with Avg
'''

df["normalized-losses"].replace(np.nan,avg1,inplace=True)

df
```

```python
'''
Now let check that we are really have no missing values in normalized-losses
'''

df["normalized-losses"].value_counts()


'''
As we can see the the value of 37 missing values in normalized-losses now are fill with average or mean of 122
'''
```

```python
'''
For practice now we are doing this for bore attribute which has 4 missing values
'''

avg2=df["bore"].astype("float").mean(axis=0)
print(avg2)
print("")
df["bore"].replace(np.nan,avg2,inplace=True)

df["bore"].value_counts()
```

```python
'''
for calculating the most commmon value for one  attribute column ==> df["..."].value_counts().idxmax()
'''

df["bore"].value_counts().idxmax()  #most common types
```

```python
'''
instead of using mean for missing values maybe we sometimes use frequency method which means using most frequent value of that column instead of that value
'''
freq=df["stroke"].value_counts().idxmax()
df["stroke"].replace(np.nan,freq,inplace=True)
df["stroke"].value_counts()
```

```python
'''
Now it is  time for data formatting we should correct the data format for some of the non nan values
'''


df[["bore","price","stroke"]]=df[["bore","price","stroke"]].astype("float")
df[["normalized-losses"]]=df[["normalized-losses"]].astype("int")
```

```python
'''
Now here we want to standard the data for instance transforming to different format or normalize that column
'''

df.head()
```

```python
# transform mpg to L/100km by mathematical operation (235 divided by mpg)
df['city-L/100km'] = 235/df["city-mpg"]

# check your transformed data 
df.head()


'''
here we add new column which is 'city-L/100km' 
'''
```

```python
'''
if we want do again the above process but without adding new column and just change that column title we can do the  below operation
'''
df["highway-mpg"] = 235/df["highway-mpg"]
# rename column name from "highway-mpg" to "highway-L/100km"
df.rename(columns={"highway-mpg":'highway-L/100km'}, inplace=True)

# check your transformed data 
df.head()
```

```python
'''
Now we are going to do normalization 
 Typical normalizations include scaling the variable so the variable average is 0,
  scaling the variable so the variable variance is 1, or scaling variable so the 
  variable values range from 0 to 1
'''

# replace (origianl value) by (original value)/(maximum value)

df["length"]=df["length"]/df["length"].max()

df["width"]=df["width"]/df["width"].max()

df[["length","width"]].head(10)

'''
Normalized version with above method for length and width
'''

```

```python

#Binning is a process of transforming continuous numerical variables into discrete categorical 'bins', for grouped analysis.

df['horsepower']=df['horsepower'].astype(float,copy=True)

binwidth=(max(df['horsepower'])-min(df['horsepower']))/4

bins=np.arange(min(df['horsepower']),max(df['horsepower']),binwidth)

bins
```

```python
groups=['low','medium','high']

#### mapping the range to these three groups
df['horsepower-levels']=pd.cut(df['horsepower'],bins,labels=groups,include_lowest=True)     

df[['horsepower','horsepower-levels']].head(30)
```

```python
'''
Now we want to visualize the bins on histogram and plot theseee bins
'''

import matplotlib.pyplot as plt
a = (0,1,2)

# draw historgram of attribute "horsepower" with bins = 3
plt.hist(df["horsepower"], bins = 3)

# set x/y labels and plot title
plt.xlabel("horsepower")
plt.ylabel("count")
plt.title("horsepower levels")

```
